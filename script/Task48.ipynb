# Aggregating data

```python
# Group then aggregate 
df.groupby('department_id').agg({'order_number':['mean']})
```

# maximum average order number is 22.902379 and minimum average order is 15.213779

# Create a loyalty flag

```python
# creating a new column for loyalty
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#remove maximum number of roles for display
pd.options.display.max_rows = None
```

```python
ords_prods_merged.head(100)
```

```python
# use the loc() function to group the customers
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
#checking the average price of product purchase by the customer groups
ords_prods_merged.groupby('loyalty_flag').agg({'prices':['mean','max','min']})
```

# prices of products differ for loyal customer (low spending habits), new customer (high spending habits) and followed by regular customer

# Spender Flag 

```python
#creating a new column for average prices for each user
ords_prods_merged['avg_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#derive new column with loc
ords_prods_merged.loc[ords_prods_merged['avg_price'] >= 10, 'spender_flag'] = 'High spender'
ords_prods_merged.loc[ords_prods_merged['avg_price'] < 10, 'spender_flag'] = 'Low spender'
```

```python
ords_prods_merged['spender_flag'].value_counts(dropna = False)
```

# Frequency Flag 

```python
#creating a new column according to median since last order
ords_prods_merged['median_order_duration'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#derive new column with loc
ords_prods_merged.loc[ords_prods_merged['median_order_duration'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merged.loc[(ords_prods_merged['median_order_duration'] > 10) & (ords_prods_merged['median_order_duration'] <= 20), 'frequency_flag'] = 'Regular customer'
ords_prods_merged.loc[ords_prods_merged['median_order_duration'] < 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merged['frequency_flag'].value_counts(dropna = False)
```

# Frequent customers are the most popular customers that purchase from the store 10 days before
